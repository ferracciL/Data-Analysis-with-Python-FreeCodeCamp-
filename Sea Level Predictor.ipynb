```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sc
```

```python
#Load data
df = pd.read_csv('epa-sea-level.csv')
df
```

```python
#Use matplotlib to create a scatter plot using the Year column as the x-axis 
#and the CSIRO Adjusted Sea Level column as the y-axis.
```

```python
plt.figure(figsize=(14,6))
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
plt.show()
```

```python
#Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. 
#Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 
#to predict the sea level rise in 2050.
```

```python
results= sc.linregress(x=df['Year'],y=df['CSIRO Adjusted Sea Level'], alternative='two-sided')
print(results)
```

```python
import numpy as np
x_pred = np.arange(df['Year'].min(), 2051)
x_pred # number of years between the first year of observation to 2050
```

```python
plt.plot((results.intercept + results.slope*x_pred))
```

```python
plt.figure(figsize=(14,6))
plt.title('rise in sea level')
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
plt.plot(x_pred,results.intercept + results.slope*x_pred,'red')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show()
```

```python
#Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset.
#Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues 
#as it has since the year 2000.
```

```python
df2000=df.loc[df.loc[:,'Year']>=2000]
df2000
```

```python
results2000= sc.linregress(x=df2000['Year'],y=df2000['CSIRO Adjusted Sea Level'], alternative='two-sided')
print(results2000)
```

```python
plt.figure(figsize=(14,6))
plt.title('rise in sea level')
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
plt.plot(x_pred,results.intercept + results.slope*x_pred,'violet',linewidth=5)
plt.plot(df2000['Year'],results2000.intercept + results2000.slope*df2000['Year'],'yellow',linewidth=5)
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show()
```
